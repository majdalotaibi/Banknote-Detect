### Banknote detect

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# load the dataset
data = pd.read_csv('Banknote.csv')
data
```

```python
v1D = data['V1']
v2D = data['V2']
```

### Statistical measures

```python
# mean
v1_mean = np.mean(v1D)
v2_mean = np.mean(v2D)

# standard deviation
v1_std = np.std(v1D)
v2_std = np.std(v2D)

# max
v1_max = np.max(v1D)
v2_max = np.max(v2D)

# min 
v1_min = np.min(v1D)
v2_min = np.min(v2D)

print('-mean:\nV1:',v1_mean,
      '\nV2:',v2_mean,
      '\n-standard deviation:\nV1:',v1_std,
       '\nV2',v2_std,
       '\n-max:\nV1:',v1_max,
       '\nV2:',v2_max,
       '\n-min:\nV1:',v1_min,
       '\nV2:',v2_min
     )
```

## Vis the data 

```python
plt.xlabel('v1')
plt.ylabel('v2')
plt.scatter(v1D, v2D)
plt.show()
```

##  K-Means clustering 

```python
V1_V2 = np.column_stack((v1D, v2D))


km_res = KMeans(n_clusters=2, n_init = 10).fit(V1_V2)
clusters = km_res.cluster_centers_
y_kmeans = km_res.predict(V1_V2)

 
plt.xlabel('V1')
plt.ylabel('V2')
plt.scatter(v1D,v2D, c=y_kmeans, cmap='inferno')
plt.scatter(clusters[:,0],clusters[:,1],s=200)

plt.show()

data['kmean'] = km_res.labels_
data['kmean'].value_counts()
```
